## WEBSCRAPE PRODUCT BASED URL

```python
##IMPORTING PACKAGES
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

## Let Scrape for Mobile MI 11x

```python
##let copy url webpage
url = 'https://www.flipkart.com/search?q=mi+11x&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_2_5_na_na_ps&otracker1=AS_QueryStore_OrganicAutoSuggest_2_5_na_na_ps&as-pos=2&as-type=RECENT&suggestionId=mi+11x%7CMobiles&requestId=37f83648-4e35-436a-9116-2d4fc4776a8f&as-backfill=on'
```

### parse content into html


```python
req = requests.get(url)
```

```python
content = BeautifulSoup(req.content, 'html.parser')
```

```python
data = content.find_all('div',{'class':'_2kHMtA'}) ## to scapre specific webpage#
```

```python
print(data)
```

```python
links = []
phn_name = []
start_link = 'https://www.flipkart.com'
```

```python
for items in data:
    rest_link=items.find('a')['href']
    name = items.find('div', attrs={'class':'_4rR01T'})
    phn_name.append(name.text)
    links.append(start_link+rest_link)
```

```python
print(phn_name[0])
print(links[0])
```

```python
dataframe = {'Phone_name':phn_name,'Links':links}
final_dataframe = pd.DataFrame(dataframe)
print(final_dataframe)
```

```python
final_dataframe.to_csv('flipkart_data.csv')
```
